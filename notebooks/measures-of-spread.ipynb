# Measures of spread

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../csv/cars.csv')
```

## Standard deviation

```python
df['price_usd'].std()
```

## Range

```python
range = df['price_usd'].max() - df['price_usd'].min()

range
```

## Quartiles

```python
median = df['price_usd'].median()
Q1 = df['price_usd'].quantile(q=0.25)
Q3 = df['price_usd'].quantile(q=0.75)

min_val = df['price_usd'].quantile(q=0)
max_val = df['price_usd'].quantile(q=1)

print(median, Q1, Q3, min_val, max_val)
```

```python
iqr = Q3 - Q1

iqr
```

### Limits for outliers detection (symmetrically distributed data)

Data between $$Q_1 -1.5 \times \text{IQR}$$ and $$Q_3 + 1.5 \times \text{IQR}$$.

```python
minlimit = Q1 - 1.5 * iqr
maxlimit = Q3 + 1.5 * iqr

print(minlimit, maxlimit)
```
