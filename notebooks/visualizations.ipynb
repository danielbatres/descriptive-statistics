```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## Scatter plot

```python
sns.scatterplot(iris, x='sepal_length', y='petal_length', hue='species')
plt.show()
```

## Joint plot

```python
sns.jointplot(iris, x='sepal_length', y='petal_length', hue='species')
plt.show()
```

## Box plot

```python
sns.boxplot(iris, x='species', y='sepal_length')
plt.show()
```

## Barplot

```python
sns.barplot(iris, x='species', y='petal_length')
plt.show()
```
